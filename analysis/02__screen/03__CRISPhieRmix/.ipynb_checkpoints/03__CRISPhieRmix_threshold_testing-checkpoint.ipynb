```r
for(name in filenames){
    print(name)
    flush.console()
    
    if (name %in% bads) {
        print("bad, skipping")
        flush.console()
        
    } else {
    data <- read.table(name, sep="\t", header=TRUE)
    
    # x needs to be the l2fcs of target genes
    tgts <- data[which(data$ctrl_status_fixed != "scramble"), ]
    x <- tgts$l2fc
    
    # geneIds needs to be the geneIds associated with x
    geneIds <- tgts$group_id_rd
    
    # negCtrl needs to be the l2fcs of scrambles
    negs <- data[which(data$ctrl_status_fixed == "scramble"), ]
    negCtrl <- negs$l2fc
    
    l2fc.CRISPhieRmix <- CRISPhieRmix(x = x, geneIds = geneIds, negCtrl = negCtrl,
                                      VERBOSE = FALSE, PLOT = FALSE)

    # all scores to df
    scores <- data.frame(groups = l2fc.CRISPhieRmix$genes, FDR = l2fc.CRISPhieRmix$FDR)

    # subset to FOXA2, SOX17, DIGIT
    ctrls <- scores[grep("FOXA2|SOX17|DIGIT|FOXD3-AS1-004", scores$groups), ]

    # write file
    file <- paste("results.", substr(name, 1, nchar(name)-4), ".txt", sep="")
    write.table(ctrls, file = file, quote = FALSE, 
                sep = "\t", row.names = FALSE)
    }
}
```
