## 5. filter sgRNAs based on count thresholds outlined above

```python
def filter_sgrnas(row):
    if row["DZ_Rep1"] >= 5 and row["DZ_Rep2"] >= 5:
        if row["Neg_Rep1"] >= 9 and row["Neg_Rep2"] >= 9:
            if row["l2fcSE"] < 1:
                return "filt"
            else:
                return "not filt (SE)"
        else:
            return "not filt (undiff)"
    else:
        return "not filt (day 0)"

data["filt_status"] = data.apply(filter_sgrnas, axis=1)
data.filt_status.value_counts()
```

```python
def filter_sgrnas_new(row):
    if row["DZ_Rep1"] >= 10.5 and row["DZ_Rep2"] >= 10.5:
        if row["Neg_Rep1"] >= 10 and row["Neg_Rep2"] >= 10:
            return "filt"
        else:
            return "not filt"
    else:
        return "not filt"

data["filt_status_new"] = data.apply(filter_sgrnas_new, axis=1)
data.filt_status_new.value_counts()
```

```python
data_filt = data[data["filt_status_new"] == "filt"]
```
