## 5. filter sgRNAs based on day zero and undifferentiated counts at various thresholds

```python
filt_threshs = list(np.linspace(0, 1000, 11))
```

```python
def filter_sgrnas(row, thresh_a, thresh_b):
    if row["DZ_Rep1"] >= thresh_a and row["DZ_Rep2"] >= thresh_a:
        if row["Neg_Rep1"] >= thresh_b and row["Neg_Rep2"] >= thresh_b:
            return "filt"
        else:
            return "not filt"
    else:
        return "not filt"
```

```python
filt_dfs = {}
filt_sgrna_nums = {}
for a in filt_threshs:
    for b in filt_threshs:
        print("a: %s // b: %s" % (a, b))
        data["filt_status"] = data.apply(filter_sgrnas, axis=1, thresh_a=a, thresh_b=b)
        data_filt = data[data["filt_status"] == "filt"]
        filt_dfs["%s__%s" % (int(a), int(b))] = data_filt
        filt_sgrna_nums["%s__%s" % (int(a), int(b))] = len(data_filt)
```

```python
filt_sgrna_nums = pd.DataFrame.from_dict(filt_sgrna_nums, orient="index").reset_index()
filt_sgrna_nums.columns = ["index", "num_sgrnas"]
filt_sgrna_nums.head()
```

## 6. write files

```python
f_dir = "../../../data/02__screen/01__normalized_counts/thresh_data"
```

```python
for key in filt_dfs:
    a = key.split("__")[0]
    b = key.split("__")[1]
    df = filt_dfs[key]
    df["group_id_rd"] = df["transcript_name"] + ";;;" + df["group_id"] + ";;;" + df["ctrl_status_fixed"]
    
    f = "%s/data_filt.DZ_%s.UD_%s.txt" % (f_dir, a, b)
    df[["group_id_rd", "ctrl_status_fixed", "l2fc"]].to_csv(f, sep="\t", index=False)
```

## 7. plot crisphie results
don't run this step until after you've run the other notebook!

```python
matrix_results = np.zeros((11, 11))

dfs = {}

for i, a in enumerate(filt_threshs):
    for j, b in enumerate(filt_threshs):

        f = "results.data_filt.DZ_%s.UD_%s.txt" % (int(a), int(b))
        
        try:
            df = pd.read_table("%s/%s" % (f_dir, f), sep="\t")
            dfs["%s__%s" % (int(a), int(b))] = df
            
            # don't consider foxd3-as1 for now
            sub = df[~df["groups"].str.contains("FOXD3-AS1")]
            
            # count n sig
            n_sig = len(sub[sub["FDR"] < 0.05])
            matrix_results[i, j] = n_sig
            
        except:
            matrix_results[i, j] = np.nan
```

```python
matrix_results = pd.DataFrame(matrix_results)
matrix_results.columns = [str(int(a)) for a in filt_threshs]
matrix_results.index = [str(int(b)) for b in filt_threshs]
matrix_results.head()
```

```python
fig = plt.figure(figsize=(3, 3))
g = sns.heatmap(matrix_results, annot=True, cmap="Blues", linewidths=1, cbar=False)
g.set_xlabel("\nUndifferentiated threshold")
g.set_ylabel("Day Zero threshold\n")
g.set_title("# controls with global CRISPhieRmix FDR < 0.05\n[proteins SOX17, FOXA2, and lncRNA DIGIT]")
```

```python
matrix_results_num = np.zeros((11, 11))

for i, a in enumerate(filt_threshs):
    for j, b in enumerate(filt_threshs):

        num = filt_sgrna_nums[filt_sgrna_nums["index"] == "%s__%s" % (int(a), int(b))]["num_sgrnas"].iloc[0]
        matrix_results_num[i, j] = num
        
matrix_results_num = pd.DataFrame(matrix_results_num)
matrix_results_num.columns = [str(int(a)) for a in filt_threshs]
matrix_results_num.index = [str(int(b)) for b in filt_threshs]
```

```python
fig = plt.figure(figsize=(3, 3))
g = sns.heatmap(matrix_results_num, annot=False, cmap="Reds", linewidths=1, cbar=True)
g.set_xlabel("\nUndifferentiated threshold")
g.set_ylabel("Day Zero threshold\n")
g.set_title("# sgRNAs that meet filter")
```
