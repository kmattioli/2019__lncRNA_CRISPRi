## examine Ihry FACS screen replicate correlations

```python
ihry_facs_f = "../../misc/10__ihry_data/TableS8.xlsx"
ihry_facs_pool1 = pd.read_excel(ihry_facs_f, sheet_name="sgRNA count pool1")
ihry_facs_pool2 = pd.read_excel(ihry_facs_f, sheet_name="sgRNA count pool2")
```

```python
ihry_growth_f = "../../misc/10__ihry_data/TableS1.xlsx"
ihry_growth_pool1 = pd.read_excel(ihry_growth_f, sheet_name="sgRNA counts POOL1")
ihry_growth_pool2 = pd.read_excel(ihry_growth_f, sheet_name="sgRNA counts POOL2")
```

```python
ihry_growth_cols = [x for x in ihry_growth_pool1.columns if x not in ["barcode", "gene_id", "symbol"]]
```

```python
ihry_facs_pool1_corr = ihry_facs_pool1[["OCT4_low_Rep1", "OCT4_low_Rep2", 
                                        "OCT4_high_Rep1", "OCT4_high_Rep2"]].corr(method="spearman")
cmap = sns.cubehelix_palette(as_cmap=True)
cg = sns.clustermap(ihry_facs_pool1_corr, figsize=(2.2, 2.2), cmap=cmap, annot=True, **{"cbar": False})
cg.cax.set_visible(False)
_ = plt.setp(cg.ax_heatmap.yaxis.get_majorticklabels(), rotation=0)
cg.savefig("Ihry_pool1_corr.pdf", dpi="figure", bbox_inches="tight")
```

```python
ihry_facs_pool2_corr = ihry_facs_pool2[["OCT4_low_Rep1", "OCT4_low_Rep2", 
                                        "OCT4_high_Rep1", "OCT4_high_Rep2"]].corr(method="spearman")
cg = sns.clustermap(ihry_facs_pool2_corr, figsize=(2.2, 2.2), cmap=cmap, annot=True, **{"cbar": False})
cg.cax.set_visible(False)
_ = plt.setp(cg.ax_heatmap.yaxis.get_majorticklabels(), rotation=0)
cg.savefig("Ihry_pool2_corr.pdf", dpi="figure", bbox_inches="tight")
```

```python
cols = ["OCT4_low_Rep1", "OCT4_low_Rep2", "OCT4_high_Rep1", "OCT4_high_Rep2"]
arrs = [np.zeros((len(cols), len(cols))), np.zeros((len(cols), len(cols)))]
count_dfs = [ihry_facs_pool1, ihry_facs_pool2]
overlap_dfs = []
for arr, counts in zip(arrs, count_dfs):
    for i, repa in enumerate(cols):
        for j, repb in enumerate(cols):
            top_a = set(counts.sort_values(by=repa, ascending=False).head(100)["barcode"])
            top_b = set(counts.sort_values(by=repb, ascending=False).head(100)["barcode"])
            a_b = len(top_a.intersection(top_b))
            arr[i, j] = int(a_b)
    df = pd.DataFrame(arr)
    df.columns = cols
    df.index = cols
    df = df.astype(int)
    overlap_dfs.append(df)
```

```python
cg = sns.clustermap(overlap_dfs[0], figsize=(2.2, 2.2), 
                    cmap="Greys", annot=True, **{"cbar": False}, fmt="g")
cg.cax.set_visible(False)
_ = plt.setp(cg.ax_heatmap.yaxis.get_majorticklabels(), rotation=0)
_ = plt.setp(cg.ax_heatmap.xaxis.get_majorticklabels(), rotation=60, ha="right", va="top")
cg.savefig("Ihry_pool1_overlap.pdf", dpi="figure", bbox_inches="tight")
```

```python
cg = sns.clustermap(overlap_dfs[1], figsize=(2.2, 2.2), 
                    cmap="Greys", annot=True, **{"cbar": False}, fmt="g")
cg.cax.set_visible(False)
_ = plt.setp(cg.ax_heatmap.yaxis.get_majorticklabels(), rotation=0)
_ = plt.setp(cg.ax_heatmap.xaxis.get_majorticklabels(), rotation=60, ha="right", va="top")
cg.savefig("Ihry_pool2_overlap.pdf", dpi="figure", bbox_inches="tight")
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
pool1_tr = np.asarray(np.log10(ihry_facs_pool1[["OCT4_low_Rep1", "OCT4_low_Rep2",
                                                "OCT4_high_Rep1", "OCT4_high_Rep2"]]+1))
print(pool1_tr.shape)

pool2_tr = np.asarray(np.log10(ihry_facs_pool2[["OCT4_low_Rep1", "OCT4_low_Rep2",
                                                "OCT4_high_Rep1", "OCT4_high_Rep2"]]+1))
print(pool2_tr.shape)
```

```python
pool1_tr_df = pd.DataFrame(data=pool1_tr)
pool1_tr_df["rowvar"] = pool1_tr_df.var(axis=1)
pool1_tr_df = pool1_tr_df.sort_values(by="rowvar", ascending=False)
pool1_top = np.asarray(pool1_tr_df.head(100)[[0, 1, 2, 3]])
print(pool1_top.shape)

pool2_tr_df = pd.DataFrame(data=pool2_tr)
pool2_tr_df["rowvar"] = pool2_tr_df.var(axis=1)
pool2_tr_df = pool2_tr_df.sort_values(by="rowvar", ascending=False)
pool2_top = np.asarray(pool2_tr_df.head(100)[[0, 1, 2, 3]])
print(pool2_top.shape)
```

```python
pool1_pca = PCA(n_components=2)
pool1_pca.fit_transform(pool1_top)
print(pool1_pca.explained_variance_ratio_)
print(pool1_pca.components_.shape)

pool1_ve = pool1_pca.explained_variance_ratio_
pool1_comp = pool1_pca.components_
pool1_comp
```

```python
pool2_pca = PCA(n_components=2)
pool2_pca.fit_transform(pool2_top)
print(pool2_pca.explained_variance_ratio_)
print(pool2_pca.components_.shape)

pool2_ve = pool2_pca.explained_variance_ratio_
pool2_comp = pool2_pca.components_
pool2_comp
```

```python
pca_pool1_df = pd.DataFrame(data=pool1_comp.T)
pca_pool1_df.columns = ["pc1", "pc2"]
pca_pool1_df["batch"] = ["rep1", "rep2", "rep1", "rep2"]
pca_pool1_df["condition"] = ["low", "low", "high", "high"]

pca_pool2_df = pd.DataFrame(data=pool2_comp.T)
pca_pool2_df.columns = ["pc1", "pc2"]
pca_pool2_df["batch"] = ["rep1", "rep2", "rep1", "rep2"]
pca_pool2_df["condition"] = ["low", "low", "high", "high"]

pal = {"low": sns.color_palette("Set2")[0], 
       "high": sns.color_palette("Set2")[1]}
```

```python
fig = plt.figure(figsize=(1.5, 1.5))
ax = sns.scatterplot(data=pca_pool1_df, x="pc1", y="pc2", hue="condition", style="batch",
                     palette=pal, s=50)
ax.set_title("Ihry OCT4 Pool #1")
plt.legend(loc=2, bbox_to_anchor=(1.01, 1))
ax.set_xlabel("PC 1 (%s%% of variance)" % round(pool1_ve[0]*100, 1))
ax.set_ylabel("PC 2 (%s%% of variance)" % round(pool1_ve[1]*100, 1))
fig.savefig("Ihry_pool1_pca.pdf", dpi="figure", bbox_inches="tight")
```

```python
fig = plt.figure(figsize=(1.5, 1.5))
ax = sns.scatterplot(data=pca_pool2_df, x="pc1", y="pc2", hue="condition", style="batch",
                     palette=pal, s=50)
ax.set_title("Ihry OCT4 Pool #2")
plt.legend(loc=2, bbox_to_anchor=(1.01, 1))
ax.set_xlabel("PC 1 (%s%% of variance)" % round(pool2_ve[0]*100, 1))
ax.set_ylabel("PC 2 (%s%% of variance)" % round(pool2_ve[1]*100, 1))
fig.savefig("Ihry_pool2_pca.pdf", dpi="figure", bbox_inches="tight")
```

```python
g = sns.jointplot(data=pool1_ctrls, x="l2fc_rep1", y="l2fc_rep2", 
                  color=sns.color_palette()[2], kind="reg",
                   joint_kws={"scatter_kws": {"s": 15, "alpha": 1, "linewidths": 0.5, 
                              "edgecolors": "white"}},
                   marginal_kws={"bins": 20, "kde": True}, size=2, space=0.5)

g.ax_joint.set_xlim(-2.5, 1)
g.ax_joint.set_ylim(-2.5, 1)

# g.ax_joint.set_xticks([-8, 1, 2, 3])
# g.ax_joint.set_yticks([0, 1, 2, 3])

g.set_axis_labels("l2fc (replicate 1)", 
                  "l2fc (replicate 2)")

# annotate number of guides above thresh
r, p = stats.pearsonr(pool1_ctrls["l2fc_rep1"], pool1_ctrls["l2fc_rep2"])
print("r: %s, p: %s" % (r, p))
r, p = stats.spearmanr(pool1_ctrls["l2fc_rep1"], pool1_ctrls["l2fc_rep2"])
g.ax_joint.text(-1.75, 1, "r = %s\np = %s\nn= %s" % ((round(r, 2), round(p, 8), len(pool1_ctrls))), 
                ha="left", va="top", fontsize=fontsize)

plt.show()
g.savefig("Ihry_controls.pdf", dpi="figure", bbox_inches="tight")
plt.close()
```
