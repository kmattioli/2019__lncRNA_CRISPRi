## 1. import data

```python
seq_counts = pd.read_table(seq_counts_f).reset_index()
seq_counts.head()
```

```python
gene_map = pd.read_table(gene_map_f, header=None)
gene_map.columns = ["gene_id", "transcript_id"]
gene_map.head()
```

```python
rna_seq = pd.read_table(rna_seq_f)
print(len(rna_seq))
rna_seq.head()
```

## 2. sep gene & transcripts

```python
gene_counts = seq_counts[seq_counts["index"].str.startswith("ENSG")]
print(len(gene_counts))
gene_counts.sample(5)
```

```python
gene_counts = gene_counts.append(seq_counts[seq_counts["index"].str.contains("DIGIT::")])
print(len(gene_counts))
```

```python
transcript_counts = seq_counts[seq_counts["index"].str.startswith("ENST")]
print(len(transcript_counts))
transcript_counts.sample(5)
```

```python
transcript_counts = transcript_counts.append(seq_counts[seq_counts["index"] == "DIGIT"])
print(len(transcript_counts))
```

## 4. make 1 dataframe with gene/transcript counts mapped to gene ids

```python
gene_counts["gene_id"] = gene_counts["index"].str.split("::", expand=True)[0]
gene_counts["type"] = "gene"
gene_counts.head()
```

```python
transcript_counts["transcript_id"] = transcript_counts["index"]
transcript_counts["type"] = "transcript"
transcript_counts.head()
```

```python
cols = ["index", "type", "hESC_rep1", "hESC_rep2", "endo_rep1", "endo_rep2", "meso_rep1", "meso_rep2"]
```

```python
tot_counts = gene_counts[cols].append(transcript_counts[cols])
tot_counts["id"] = tot_counts["index"].str.split("::", expand=True)[0]
tot_counts.sample(5)
```

```python
tot_counts = tot_counts.merge(gene_map, left_on="id", right_on="transcript_id", how="left")
tot_counts.sample(5)
```

```python
def fill_gene_id(row):
    if pd.isnull(row["gene_id"]):
        return row["id"]
    else:
        return row["gene_id"]
    
tot_counts["gene_id"] = tot_counts.apply(fill_gene_id, axis=1)
tot_counts.sample(5)
```

```python
tot_counts = tot_counts[["id", "gene_id", "type", "hESC_rep1", "hESC_rep2", "endo_rep1", "endo_rep2",
                        "meso_rep1", "meso_rep2"]]
tot_counts = tot_counts.sort_values(by="gene_id")
tot_counts.head()
```

## 5. sum counts per gene (incl. gene isoform and all transcripts)

```python
sum_counts = tot_counts.groupby("gene_id")[["hESC_rep1", "hESC_rep2", "endo_rep1", "endo_rep2",
                                            "meso_rep1", "meso_rep2"]].agg("sum").reset_index()
sum_counts.head()
```

## 6. calculate splicing efficiency

```python
len(gene_counts)
```

```python
len(sum_counts)
```

```python
data = gene_counts.merge(sum_counts, on="gene_id", suffixes=("_gene", "_tot"))
data.head()
```

```python
data["hESC_rep1_eff"] = 1-data["hESC_rep1_gene"]/data["hESC_rep1_tot"]
data["hESC_rep2_eff"] = 1-data["hESC_rep2_gene"]/data["hESC_rep2_tot"]
data["endo_rep1_eff"] = 1-data["endo_rep1_gene"]/data["endo_rep1_tot"]
data["endo_rep2_eff"] = 1-data["endo_rep2_gene"]/data["endo_rep2_tot"]
data["meso_rep1_eff"] = 1-data["meso_rep1_gene"]/data["meso_rep1_tot"]
data["meso_rep2_eff"] = 1-data["meso_rep2_gene"]/data["meso_rep2_tot"]
```

```python
data["hESC_eff"] = data[["hESC_rep1_eff", "hESC_rep2_eff"]].mean(axis=1)
data["endo_eff"] = data[["endo_rep1_eff", "endo_rep2_eff"]].mean(axis=1)
data["meso_eff"] = data[["meso_rep1_eff", "meso_rep2_eff"]].mean(axis=1)
```

```python
data["hESC_exp"] = data[["hESC_rep1_tot", "hESC_rep2_tot"]].mean(axis=1)
data["endo_exp"] = data[["endo_rep1_tot", "endo_rep2_tot"]].mean(axis=1)
data["meso_exp"] = data[["meso_rep1_tot", "meso_rep2_tot"]].mean(axis=1)
```

## 7. make final dataframe

```python
data = data[["gene_id", "hESC_eff", "endo_eff", "meso_eff", "hESC_exp", "endo_exp", "meso_exp"]]
data.head()
```

```python
hESC_exp = data[data["hESC_exp"] >= 1]
hESC_exp.sort_values(by="hESC_eff", ascending=False).head()
```
